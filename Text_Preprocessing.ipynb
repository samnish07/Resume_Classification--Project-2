#### <font color='purple'><b>Importing libraries :

```python
import pandas as pd
import re
import string
from nltk.corpus import stopwords
from collections import Counter
from nltk import word_tokenize
from nltk.util import ngrams 
```

#### <font color='purple'><b>Read Dataset :

```python
df = pd.read_csv('resumes.csv')
df.head()
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
df.head()
```

### <font color='magenta'><b>Text Cleaning :

```python
df1 = df.copy()
df1.drop(['category'],axis=1,inplace=True)
```

#### <font color='maroon'><b>Removing the newline, tab and carriage return characters \n \r \t :

```python
df1['clean_text'] = df1['resume_text'].replace(r'\r+|\n+|\t+','',regex=True)
df1.head()
```

#### <font color='maroon'><b>Removing urls :

```python
def remove_url(text):
    return re.sub(r'https?://\S+|www\.\S+', '',text)
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_url(x))
df1.head()
```

#### <font color='maroon'><b>Convert to lowercase :

```python
df1['clean_text'] = df1['clean_text'].str.lower()
df1.head()
```

#### <font color='maroon'><b>Removal of Punctuations :

```python
string.punctuation
```

```python
def remove_punctuations(text):
    punctuations = string.punctuation
    return text.translate(str.maketrans('','',punctuations))
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_punctuations(x))
df1.head(10)
```

#### <font color='maroon'><b>Removal of Stopwords :

```python
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in STOPWORDS])
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_stopwords(x))
df1.head(10)
```

#### <font color='maroon'><b>Removal of Least Frequent words :

```python
word_count = Counter()
for text in df1['clean_text']:
    for word in text.split():
        word_count[word] += 1
        
#word_count.most_common(50)       
```

<font color='orange'><b><I>Finding words that occured only once :

```python
for w in word_count.most_common():
    if w[1]<2:
        RARE_WORDS = w
        #print(RARE_WORDS)
```

```python
def remove_rare_words(text):
    return " ".join([word for word in text.split() if word not in RARE_WORDS])
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_rare_words(x))
df1.head(10)
```

#### <font color='maroon'><b>Removal of Special Characters and Additional Spaces :

```python
def remove_spl_chars(text):
    text = re.sub('[^a-zA-Z0-9]'," ",text)
    text = re.sub('\s+',' ',text)
    return text
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_spl_chars(x))
df1.head(20)
```

```python
df1['clean_text'][4]
```

<font color='orange'><b><I>Saving the cleaned resume text in .csv format :

```python
df1.to_csv('resume_cleantext.csv')
```

### <font color='magenta'><b>Tokenization :

<font color='hotpink'><b><I>Tokenization is to split the text into smaller units called tokens, separated by commas.

```python
df1['tokenized_text'] = df1['clean_text'].apply(lambda x: word_tokenize(x))
df1.drop(['resume_text'],axis=1,inplace=True)
df1.head(10)
```

### <font color='purple'><b>Extract Skills required for the Resume Classification :

```python
keywords = ['react', 'react developer','reactjs', 'reactjs developer', 'peoplesoft', 'peoplesoft dba','peoplesoft admin',
            'workday','workday consultant', 'hcm', 'eib', 'workday hcm', 'workday studio', 'sql', 'oracle sql', 'mysql', 
            'sql developer', 'dba', 'db','oracle','admin', 'experience']
```

```python
df2 = df1.copy()
df2.head()
```

<font color='orange'><b><I> Function to extract one-gram words in the keywords for skills

```python
def skillSet(text):
    t = []
    for i in range(len(text)):
        if text[i] in keywords:
            if text[i] in t:
                continue
            t.append(text[i]) 
    return(t)
```

```python
df2['skillset'] = df2['tokenized_text'].apply(lambda x: skillSet(x))
df2.head(20)
```

```python
df3 = df2.copy()
```

```python
df3.drop(['clean_text','skillset'],axis=1,inplace=True)
```

```python
df3.head()
```

### <font color='maroon'><b>Finding N-grams :

<font color='brown'><b><I>Tokens of any number of consecutive written words is known as Ngram.

```python
def extract_ngrams(data,num):
    n_grams = ngrams(data,num)
    return[' '.join(grams) for grams in n_grams]
```

```python
df3['bigram'] = df3['tokenized_text'].apply(lambda x: extract_ngrams(x,2))
df3.head(10)
```

<font color='orange'><b><I> Function to extract bigram words in the keywords for skills :

```python
def skillSet(text):
    t = []
    for i in range(len(text)):
        if text[i].strip() in keywords:
            if text[i] in t:
                continue
            t.append(text[i]) 
    return(t)
```

```python
df3['skillset2'] = df3['bigram'].apply(lambda x: skillSet(x))
df3.head(50)
```

```python
df_skill = df.copy()
df_skill.drop(['resume_text'],axis=1,inplace=True)
df_skill['resume_skills'] = df2['skillset'] + df3['skillset2']
df_skill = df_skill.iloc[:,[1,0]]
df_skill.head()
```

### <font color='purple'><b>Final Dataframe with the keywords of Skills in the resume and the Categories :

```python
df_clean = df_skill.copy()
tokenized = df_skill['resume_skills']
for i, tokens in enumerate(tokenized):
    tokenized[i] = ' '.join(tokens)

df_clean['skills'] = tokenized
df_clean.drop(['resume_skills'],axis=1,inplace=True)
df_clean = df_clean.iloc[:,[1,0]]
df_clean.head(10)
```

<font color='orange'><b><I>Saving the dataset in .csv format :

```python
df_clean.to_csv("resume_skills.csv")
```
