#### <font color='purple'><b>Importing libraries :

```python
import pandas as pd
import re
import string
from nltk.corpus import stopwords
from collections import Counter
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
```

#### <font color='purple'><b>Read Dataset :

```python
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['resumes.csv']))
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
df.head()
```

### <font color='magenta'><b>Text Cleaning :

```python
df1 = df.copy()
df1.drop(['category'],axis=1,inplace=True)
```

#### <font color='maroon'><b>Removing the newline, tab and carriage return characters \n \r \t :

```python
df1['clean_text'] = df1['resume_text'].replace(r'\r+|\n+|\t+','',regex=True)
df1.head()
```

#### <font color='maroon'><b>Removing urls :

```python
def remove_url(text):
    return re.sub(r'https?://\S+|www\.\S+', '',text)
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_url(x))
df1.head()
```

#### <font color='maroon'><b>Convert to lowercase :

```python
df1['clean_text'] = df1['clean_text'].str.lower()
df1.head()
```

#### <font color='maroon'><b>Removal of Punctuations :

```python
string.punctuation
```

```python
def remove_punctuations(text):
    punctuations = string.punctuation
    return text.translate(str.maketrans('','',punctuations))
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_punctuations(x))
df1.head(10)
```

#### <font color='maroon'><b>Removal of Stopwords :

```python
import nltk
```

```python
 nltk.download('stopwords')
```

```python
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in STOPWORDS])
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_stopwords(x))
df1.head(10)
```

#### <font color='maroon'><b>Removal of Least Frequent words :

```python
word_count = Counter()
for text in df1['clean_text']:
    for word in text.split():
        word_count[word] += 1
        
#word_count.most_common(50) 
```

<font color='orange'><b><I>Finding words that occured only once :

```python
for w in word_count.most_common():
    if w[1]<2:
        RARE_WORDS = w
        #print(RARE_WORDS)
```

```python
def remove_rare_words(text):
    return " ".join([word for word in text.split() if word not in RARE_WORDS])
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_rare_words(x))
df1.head(10)
```

#### <font color='maroon'><b>Removal of Special Characters and Additional Spaces :

```python
def remove_spl_chars(text):
    text = re.sub('[^a-zA-Z0-9]'," ",text)
    text = re.sub('\s+',' ',text)
    return text
```

```python
df1['clean_text'] = df1['clean_text'].apply(lambda x: remove_spl_chars(x))
df1.head(20)
```

```python
df1['clean_text'][4]
```

### <font color='magenta'><b>Tokenization :

```python
nltk.download('punkt')
```

```python
df1['tokenized_text'] = df1['clean_text'].apply(lambda x: word_tokenize(x))
df1.drop(['resume_text'],axis=1,inplace=True)
df1.head(10)
```

### <font color='magenta'><b>Lemmatization :

```python
word_lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
nltk.download('omw-1.4')
```

```python
df1['lemmatized_text'] = df1['tokenized_text'].apply(lambda x: [word_lemmatizer.lemmatize(i) for i in x])
df1.head(10)
```

### <font color='purple'><b>Final Dataframe :

```python
df_clean = df1.copy()
#df_clean.drop(['#'])
tokenized = df_clean['lemmatized_text']
for i, tokens in enumerate(tokenized):
    tokenized[i] = ' '.join(tokens)

df_clean.drop(['clean_text','tokenized_text'],axis=1,inplace=True)
df_clean = df_clean.join(df['category'])
df_clean.rename(columns={'lemmatized_text':'clean_text'},inplace=True)
df_clean.head()
```

<font color='orange'><b><I>Saving it in .csv format :

```python
df_clean.to_csv("resume_clean.csv",index=False,encoding='utf-8-sig')
files.download('resume_clean.csv')
```
